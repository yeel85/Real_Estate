```python
from urlparse import urlparse
import pandas as pd
import requests
import html5lib
```

```python
o = requests.get('https://www.realestate.com.au/auction-results/vic?rsf=edm:auction:vic&appRedirection=true')
```

```python
content_ =o.content
df_ = pd.read_html(content_)
```

```python
suburb_ = {}
for sub_ in range(len(df_)):
#     x['location'] = str(a)
    temp = df_[sub_]
    a = temp.columns[0].split()
    temp['location'] = str(a)
    temp.columns.values[0] = 'Place'   
    suburb_[sub_] = temp
    temp = []
results = pd.concat(suburb_)
#     print(df_[sub_])
```

```python
# results = results.set_index('Place')
results = results.reset_index()
results = results.drop(['level_0','level_1'],axis=1)
```

```python
results = results.rename(columns = {'\n                      Agency\n                    ':'Agency'})
results = results.rename(columns = {'\n                      Beds\n                    ':'Bed'})
results = results.rename(columns = {'\n                      Price\n                    ':'Price'})
results = results.rename(columns = {'\n                      Result\n                    ':'Result'})
results = results.rename(columns = {'\n                      Date\n                    ':'Date'})
results = results.rename(columns = {'\n                      Type\n                    ':'Type'})
```

```python
results["Price"] = (results['Price'].replace( '[\$,)]','', regex=True ))
results.set_index('Place')
```

```python
import decimal

pricing_ = []
for i in range(len(results)):
    if results["Price"][i] == u'-' :
        pricing_.append("NaN")
    else:
        pricing_.append(float(decimal.Decimal(results["Price"][i])))
results['Cost'] = pricing_
```
